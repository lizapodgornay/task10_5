```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

def website_analyse (dfile):
    df = pd.read_csv(dfile)

    # вычисление среднего арифметического
    mean = np.mean(df['Просмотренные страницы'])
    print(f'Среднее количество просмотренных страниц: {mean}')

    # вычисление стандартного отклонения
    std = np.std(df['Время на сайте (сек)'])
    print(f'Стандартное отклонение: {std}')

    # коэффициент вариации
    mean_time = np.mean(df['Время на сайте (сек)'])
    variation = std/mean_time
    if variation > 0.25:
        print(f'Высокая вариабельность: {variation}')
    elif 0.1 < variation < 0.25:
        print(f'Умеренная вариабельность: {variation}')
    else:
        print(f'Слабая вариабельность: {variation}')

    # фильтрация столбцов Тип устройства
    filtered_mobile = df[df['Тип устройства'] == 'Мобильное']['Просмотренные страницы']
    filtered_des = df[df['Тип устройства'] == 'Десктоп']['Просмотренные страницы']

    # статистически значимая разница
    stat, p_value = ttest_ind(filtered_mobile, filtered_des)
    if p_value > 0.05:
        print(f'Есть статистически значимая разница в количестве страниц {p_value}')
    else:
        print(f'Нет статистически значимой разницы: {p_value}')
```
